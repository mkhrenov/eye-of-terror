```julia
using ClosedHeaven
```

```julia
motor = ClosedHeaven.AxialFluxMotor(
    10,         # Number of poles
    12,         # Number of slots (8 per phase)
    1,          # Number of turns per slot
    200e-3,     # Stator winding OD (m)
    150e-3,        # Stator winding ID (m)
    0.3,        # Magnetic field strength, (Tesla) (BX082)
    2.052e-3,    # Winding wire diameter (16 AWG)
    1.68e-8     # Winding wire resistivity (Ωm)
)
```

```julia
ke = ClosedHeaven.ke(motor) # V / (rad/s)
kt = ClosedHeaven.kt(motor) # Nm / A
Rm = ClosedHeaven.resistance(motor) # Ω
J =  7842607.63 * 1e-9 # kg m²
```

```julia
@show Rm
```

```julia
@show 1 / (7500 / (6 * 3.7) * 2π / 60)
@show ke

@show ke * kt / (J * Rm)
@show 7000 / 60 * 2π
```
