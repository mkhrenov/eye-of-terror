```julia
using ClosedHeaven
```

```julia
motor = ClosedHeaven.AxialFluxMotor(
    20,         # Number of poles
    21,         # Number of slots (8 per phase)
    4,          # Number of turns per slot
    100e-3,     # Stator winding OR (m)
    87e-3,      # Stator winding IR (m)
    0.3,        # Magnetic field strength, (Tesla) (BX082)
    1.628e-3,   # Winding wire diameter (14 AWG)
    1.68e-8     # Winding wire resistivity (Ωm)
)
```

```julia
ke = ClosedHeaven.ke(motor) # V / (rad/s)
kt = ClosedHeaven.kt(motor) # Nm / A
Rm = ClosedHeaven.resistance(motor) # Ω
J =  7842607.63 * 1e-9 # kg m²
```

```julia
@show Rm
```

```julia
@show 1 / (7500 / (6 * 3.7) * 2π / 60)
@show ke

@show ke * kt / (J * Rm)
@show 7000 / 60 * 2π
```
